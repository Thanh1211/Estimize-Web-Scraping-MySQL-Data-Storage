```python
# Get a random value between 8 and 17 for light sleep
def random_sleep():
    timer = np.random.randint(8,17)
    time.sleep(timer)


```

```python
 
chrome_options = webdriver.ChromeOptions()
chrome_options.headless = False
browser = webdriver.Chrome(executable_path = "C:\Program Files\ChromeDriver\chromedriver.exe", options = chrome_options)
  

login_link = 'https://www.estimize.com/users/sign_in'
browser.get(login_link)

random_sleep()
username = '*****'
password = '*****'

user_field = browser.find_element(By.NAME,"user[login]")
user_field.send_keys(username)
        
password_field = browser.find_element(By.NAME,("user[password]"))
password_field.send_keys(password)

random_sleep()

btn = browser.find_element(By.XPATH,("//input[@type='submit' and @value='Log In']"))
btn.click()
```

```python
TICKER_LIST=['AMZN', 'AAPL', 'MSFT', 'GOOG', 'TSLA', 'JNJ', 'PG', 'NVDA', 'CSCO', 'BABA', 'HD',
'BIDU', 'WMT', 'CRM', 'LULU', 'TGT', 'PANW', 'ADBE', 'VMW', 'MU', 'NKE', 'ORCL', 'BB', 'HPQ', 'COST', 'AMAT',
'BAC', 'CVX', 'AMGN']
```

```python
Companies={}
for i in range(len(TICKER_LIST)):
    url ='https://www.estimize.com/'+str(TICKER_LIST[i])+'/fq1-2020?metric_name=eps&chart=table' 
    browser.get(url)
    html=browser.page_source
    soup = BeautifulSoup(html,'lxml')
    random_sleep()
    Company={}
    #companies basic information
    Company['Ticker'] = soup.find('h1', {'class':"release-header-information-title"}).text
    Company['Name'] = soup.find('p', {'class':"release-header-information-description"}).text
    Company['Sectors'] = browser.find_element(By.XPATH,('//*[@class = "release-header-information-breadcrumb"]/span/a[1]/span')).text
    Company['Industry'] = browser.find_element(By.XPATH,('//*[@class = "release-header-information-breadcrumb"]/span[2]/a[1]/span')).text
    Company['Follower'] = soup.find('div', {'class':'estimize-row summary-sub-title'}).text
    Company['Analyst'] = soup.find('a', {'class':"estimize-row summary-sub-title"}).text
    Companies[i] = Company

df = pd.DataFrame(data=Companies)
stacked = df.stack()
df2 = stacked.unstack(0)
df2
```

```python

df2.to_sql(con= database, name='companies', if_exists='append',index=False)
```

```python
estimize_eps1={}
for x in range(len(TICKER_LIST1)):
    for y in range(1,5):
        url=f'https://www.estimize.com/'+str(TICKER_LIST1[x])+'/fq'+str(y)+'-2020?metric_name=eps&chart=historical'
        browser.get(url)
        html=browser.page_source
        soup = BeautifulSoup(html,'lxml')
        random_sleep()  
        eps={}
        #companies basic information
        eps['Quarter'] =browser.find_element(By.XPATH,'//*[@id="main-content"]/div[2]/div[3]/h2[2]/span[2]').text
        eps['Ticker'] = soup.find('h1', {'class':"release-header-information-title"}).text
        eps['Actual'] = browser.find_element(By.XPATH,('//*[@id="estimates"]/table/tbody[1]/tr[1]/td[5]')).text
        eps['Estimize Consensus'] = browser.find_element(By.XPATH,('//*[@id="estimates"]/table/tbody[1]/tr[2]/td[5]')).text
        eps['Mean'] = browser.find_element(By.XPATH,('//*[@id="estimates"]/table/tbody[1]/tr[3]/td[5]')).text
        eps['Wall Street'] = browser.find_element(By.XPATH,('//*[@id="estimates"]/table/tbody[1]/tr[4]/td[5]')).text
        estimize_eps1[x,y] = eps

```
