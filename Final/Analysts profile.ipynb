```python
df=pd.read_csv('Analysts_list.csv', skiprows=0)
df2=df['Analyst Id'].values.tolist()

analyst_id=df2[1:8896]
analyst_id
```

```python
random_sleep()
analysts = []
for x in analyst_id:
        url='https://www.estimize.com/users/'+str(x)
        browser.get(url)#get link
        random_sleep()
    #find element and scroll
        try:
            users = {}
            users['Analyst Name'] = browser.find_element(By.XPATH,('//*[@id="users_show"]/div[4]/div[1]/div[1]/div[1]/div/h1/a')).text
            users['roles'] = browser.find_element(By.XPATH,('//*[@id="users_show"]/div[4]/div[1]/div[1]/div[1]/div/ul')).text
            users['Join_date'] =  browser.find_element(By.XPATH,('//*[@id="users_show"]/div[4]/div[1]/div[1]/div[1]/div/div[2]/div[2]')).text
            users['Confidence_score'] =  browser.find_element(By.XPATH,('//*[@id="confidence-wrap"]/div/div[2]')).text
            users['error_rate'] =  browser.find_element(By.XPATH,('//*[@id="profile-tab-wrap"]/div[1]/div[1]/div[1]')).text
            users['accuracy_percentile'] =  browser.find_element(By.XPATH,('//*[@id="profile-tab-wrap"]/div[1]/div[2]/div[1]')).text
            users['points'] =  browser.find_element(By.XPATH,('//*[@id="profile-tab-wrap"]/div[2]/div[1]/div[1]')).text
            users['points/estimate'] =  browser.find_element(By.XPATH,('//*[@id="profile-tab-wrap"]/div[2]/div[2]/div[1]')).text
            users['stocks'] =  browser.find_element(By.XPATH,('//*[@id="profile-tab-wrap"]/div[3]/div[1]/div[1]')).text
            users['pending'] =  browser.find_element(By.XPATH,('//*[@id="profile-tab-wrap"]/div[3]/div[2]/div[1]')).text
            analysts.append(users)
            print('analyst #'+str(df2.index(x)) +' - ' + str(x) + " data sucessfully retrived")
        except NoSuchElementException:
            print('analyst #'+str(df2.index(x)) +' - ' + str(x) + " did not have data")
            pass

print('this batch is complete!')        
```

```python
dfanlysts= pd.DataFrame(analysts) 
dfanlysts['Join_date'] = dfanlysts['Join_date'].str.replace(' - .*', '')
dfanlysts['Join_date'] = dfanlysts['Join_date'].str.replace('Member since ', '')
dfanlysts
```

```python
dfanlysts.to_sql(con= database, name='analysts_profile', if_exists='append',index=False)
```
