```python
# Get a random value between 8 and 17 for light sleep
def random_sleep():
    timer = np.random.randint(8,17)
    time.sleep(timer)
```

```python
chrome_options = webdriver.ChromeOptions()
chrome_options.headless = False
s = Service("C:\Program Files\ChromeDriver\chromedriver.exe")
browser = webdriver.Chrome(service=s)
  

login_link = 'https://www.estimize.com/users/sign_in'
browser.get(login_link)

random_sleep()
username = 'arianne.de.w@gmail.com'
password = '@Data1234'

user_field = browser.find_element(By.NAME,"user[login]")
user_field.send_keys(username)
        
password_field = browser.find_element(By.NAME,("user[password]"))
password_field.send_keys(password)

random_sleep()

btn = browser.find_element(By.XPATH,("//input[@type='submit' and @value='Log In']"))
btn.click()
```

```python
TICKER_LIST=['MU', 'NKE', 'ORCL', 'BB']


#TICKER_LIST=[ 'HPQ', 'COST', 'AMAT','BAC', 'CVX', 'AMGN']
```

```python
#dictionary for data in for
random_sleep()
analyst = []
for x in TICKER_LIST:
    for y in range(1,5):
        url=f'https://www.estimize.com/'+str(x)+'/fq'+str(y)+'-2020?metric_name=eps&chart=historical'
        browser.get(url)#get link
        random_sleep()
    #find element and scroll
        try:
            random_sleep()
            browser.execute_script("window.scrollTo(0, 3000)")# Scroll to table # pause 
            browser.find_element(By.XPATH,('/html/body/div[4]/div[2]/div/div[2]/div[2]/div[3]/div[4]/div/a')).click()# click to reveal.
        except:
            pass
        a = int(browser.find_element(By.XPATH,('//*[@id="estimates"]/p[1]/strong/span')).text)+1
        random_sleep()
        for z in range(1,a):
            try:
                datas = {}
                datas['Ticker'] = x
                datas['Quarter'] = 'Q' + str(y)+ ' 2020'
                datas['Analyst Name'] = browser.find_element(By.XPATH,('//*[@id="estimates"]/table/tbody[2]/tr['+str(z)+']/td[3]/a[2]/strong')).text
                datas['Analyst Id'] = browser.find_element(By.XPATH,('//*[@id="estimates"]/table/tbody[2]/tr['+str(z)+']/td[3]/a[2]/strong[@data-user]')).get_attribute("data-user")
                datas['Analyst Rank'] =  browser.find_element(By.XPATH,('//*[@id="estimates"]/table/tbody[2]/tr['+str(z)+']/td[4]')).text
                datas['Estimate'] =  browser.find_element(By.XPATH,('//*[@id="estimates"]/table/tbody[2]/tr['+str(z)+']/td[6]')).text
                datas['Analyst Confidence'] =  browser.find_element(By.XPATH,('//*[@id="estimates"]/table/tbody[2]/tr['+str(z)+']/td[7]')).text
                analyst.append(datas)
            except:
                continue
```

```python
df= pd.DataFrame(analyst) 
df

```

```python
df.to_sql(con= database, name='analyst_covered_stock_20', if_exists='append',index=False)
```
